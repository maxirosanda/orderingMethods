```python
# ESTRUCTURA Empleado
class Empleado:
    def __init__(self, numero_empleado, nombre):
        self.numero_empleado = numero_empleado
        self.nombre = nombre
        self.sueldos_mensuales = [0.0] * 12  # Sueldos mensuales
        self.sueldo_anual = 0.0
        self.porcentaje_sueldo_mes = [0.0] * 12  # Porcentaje del sueldo por mes

# Variables globales
empleados = [Empleado(i, f"Empleado_{i + 1}") for i in range(30)]
total_sueldos_mensuales = [0.0] * 12
max_sueldo_mes = [0.0] * 12
nombre_empleado_menor_sueldo_primer_semestre = ""
meses_sueldos_totales_menores_4millones = 0

def leer_nomina():
    global empleados
    with open("nomina.txt", "r") as file:
        for i in range(30):
            datos = file.readline().strip().split(",")
            empleados[i].numero_empleado = int(datos[0])
            empleados[i].nombre = datos[1]
            for j in range(12):
                empleados[i].sueldos_mensuales[j] = float(datos[j + 2]) * int(datos[14 + j])  # Valor hora * Cantidad horas

    print("Empleados cargados correctamente desde archivo nomina")

def calcular_sueldos_anuales_por_empleado():
    for empleado in empleados:
        empleado.sueldo_anual = sum(empleado.sueldos_mensuales)
        print(f"Empleado: {empleado.nombre} sueldo total anual: {empleado.sueldo_anual}")

def calcular_total_sueldos_mensuales():
    for j in range(12):
        total_sueldos_mensuales[j] = sum(empleado.sueldos_mensuales[j] for empleado in empleados)
        print(f"Mes: {j + 1} Total mensual: {total_sueldos_mensuales[j]}")

def calcular_max_sueldo_mes():
    for j in range(12):
        max_sueldo_mes[j] = max(empleado.sueldos_mensuales[j] for empleado in empleados)
        print(f"Máximo sueldo mes {j + 1}: {max_sueldo_mes[j]}")

def calcular_porcentaje_sueldo_mes():
    for empleado in empleados:
        for j in range(12):
            porcentaje = (empleado.sueldos_mensuales[j] / total_sueldos_mensuales[j]) * 100 if total_sueldos_mensuales[j] > 0 else 0
            empleado.porcentaje_sueldo_mes[j] = porcentaje
            print(f"Empleado: {empleado.nombre} - Porcentaje: {porcentaje}% - Mes: {j + 1}")

def determinar_empleado_menor_sueldo_primer_semestre():
    global nombre_empleado_menor_sueldo_primer_semestre
    menor_sueldo = float('inf')
    for empleado in empleados:
        for j in range(6):  # Primer semestre (meses 0 a 5)
            if empleado.sueldos_mensuales[j] < menor_sueldo:
                menor_sueldo = empleado.sueldos_mensuales[j]
                nombre_empleado_menor_sueldo_primer_semestre = empleado.nombre

    print(f"Empleado con menor sueldo en el primer semestre: {nombre_empleado_menor_sueldo_primer_semestre}")

def calcular_meses_sueldos_totales_menores_4millones():
    global meses_sueldos_totales_menores_4millones
    meses_sueldos_totales_menores_4millones = 0
    for j in range(12):
        if total_sueldos_mensuales[j] < 4000000:
            meses_sueldos_totales_menores_4millones += 1
    print(f"Meses con sueldos totales menores a $4 millones: {meses_sueldos_totales_menores_4millones}")

def guardar_liquidacion():
    with open("liquidacion.txt", "w") as file:
        empleados.sort(key=lambda e: e.sueldo_anual, reverse=True)  # Ordenar empleados por sueldo anual
        for empleado in empleados:
            file.write(f"{empleado.nombre},{empleado.sueldo_anual}\n")

    print("Liquidación guardada correctamente")

# Llamadas a funciones (Ejemplo de uso)
leer_nomina()
calcular_sueldos_anuales_por_empleado()
calcular_total_sueldos_mensuales()
calcular_max_sueldo_mes()
calcular_porcentaje_sueldo_mes()
determinar_empleado_menor_sueldo_primer_semestre()
calcular_meses_sueldos_totales_menores_4millones()
guardar_liquidacion()
```
