```python
class Letra:
    def __init__(self, letra):
        self.letra = letra
        self.siguiente = None

class Revista:
    def __init__(self, nombre="", pagina="", letra_frontal=None, letra_trasera=None):
        self.nombre = nombre
        self.pagina = pagina
        self.letra_frontal = letra_frontal
        self.letra_trasera = letra_trasera

class RevistaEncontrada:
    def __init__(self, nombre="", pagina="", letra=None):
        self.nombre = nombre
        self.pagina = pagina
        self.letra = letra

class Venta:
    def __init__(self, nombre_cliente="", revista_comprada=""):
        self.nombre_cliente = nombre_cliente
        self.revista_comprada = revista_comprada

letras = None

def insertar_letra_lista(lista, letra):
    nuevo_nodo = Letra(letra)
    if lista is None:
        return nuevo_nodo
    else:
        nodo_actual = lista
        while nodo_actual.siguiente:
            nodo_actual = nodo_actual.siguiente
        nodo_actual.siguiente = nuevo_nodo
        return lista

letras = insertar_letra_lista(letras, "Arial12pxRojo")
letras = insertar_letra_lista(letras, "Times14pxAzul")
letras = insertar_letra_lista(letras, "Helvetica16pxVerde")
letras = insertar_letra_lista(letras, "Courier10pxNegro")
letras = insertar_letra_lista(letras, "Verdana18pxNaranja")
letras = insertar_letra_lista(letras, "ComicSans12pxVioleta")
letras = insertar_letra_lista(letras, "Georgia14pxMarrón")
letras = insertar_letra_lista(letras, "Arial16pxGris")

revistas = [Revista() for _ in range(50)]
revistas_contienen_letras = [RevistaEncontrada() for _ in range(50)]
ventas = [Venta() for _ in range(50)]

def leer_revistas():
    with open("revistas.txt", "r") as file:
        for i in range(50):
            linea = file.readline().strip()
            if linea:
                datos = linea.split(",")
                revistas[i].nombre = datos[0]
                revistas[i].pagina = datos[1]
                agregar_letra_cola(revistas[i], datos[2])

def leer_historial_ventas():
    with open("ventas.txt", "r") as file:
        for i in range(50):
            linea = file.readline().strip()
            if linea:
                datos = linea.split(",")
                ventas[i].revista_comprada = datos[0]
                ventas[i].nombre_cliente = datos[1]

def agregar_letra_cola(revista, letra):
    nueva_letra = Letra(letra)
    if revista.letra_frontal is None:
        revista.letra_frontal = nueva_letra
        revista.letra_trasera = nueva_letra
    else:
        revista.letra_trasera.siguiente = nueva_letra
        revista.letra_trasera = nueva_letra

def traer_letra_cola(revista):
    if revista.letra_frontal:
        letra_actual = revista.letra_frontal.letra
        revista.letra_frontal = revista.letra_frontal.siguiente
        return letra_actual
    return None

def buscar_lista(lista, posicion):
    nodo_actual = lista
    for i in range(1, posicion + 1):
        if nodo_actual is None:
            return None
        if i == posicion:
            return nodo_actual.letra
        nodo_actual = nodo_actual.siguiente
    return None

def principal():
    for i in range(50):
        while revistas[i].letra_frontal is not None:
            for j in range(1, 9):
                letra = buscar_lista(letras, j)
                if letra == revistas[i].letra_frontal.letra:
                    revistas_contienen_letras[j - 1].nombre = revistas[i].nombre
                    revistas_contienen_letras[j - 1].pagina = revistas[i].pagina
                    revistas_contienen_letras[j - 1].letra = revistas[i].letra_frontal
            traer_letra_cola(revistas[i])

    for j in range(50):
        for i in range(50):
            if revistas_contienen_letras[j].nombre == ventas[i].revista_comprada:
                print(f"Posible coincidencia: Cliente {ventas[i].nombre_cliente} compró la revista {ventas[i].revista_comprada}")

# Ejecución de las funciones de lectura
leer_revistas()
leer_historial_ventas()

# Llamada a la función principal
principal()
```
