```python
class Letra:
    def __init__(self, letra):
        self.letra = letra
        self.siguiente = None

class Revista:
    def __init__(self, nombre, pagina):
        self.nombre = nombre
        self.pagina = pagina
        self.letra_frontal = None
        self.letra_trasera = None

class RevistaEncontrada:
    def __init__(self, nombre, pagina, letra):
        self.nombre = nombre
        self.pagina = pagina
        self.letra = letra

class Venta:
    def __init__(self, nombre_cliente, revista_comprada):
        self.nombre_cliente = nombre_cliente
        self.revista_comprada = revista_comprada

def insertar_letra_lista(lista, letra):
    nuevo_nodo = Letra(letra)
    
    if lista is None:
        return nuevo_nodo
    
    # Inserta al inicio de la lista
    nuevo_nodo.siguiente = lista
    return nuevo_nodo

def agregar_letra(letra_frontal, letra_trasera, letra):
    nueva_letra = Letra(letra)
    
    if letra_frontal is None:
        letra_frontal = nueva_letra
        letra_trasera = nueva_letra
    else:
        letra_trasera.siguiente = nueva_letra
        letra_trasera = nueva_letra
    return letra_frontal, letra_trasera

def traer_letra(letra_frontal):
    if letra_frontal is not None:
        letra_actual = letra_frontal.letra
        letra_frontal = letra_frontal.siguiente
        return letra_actual, letra_frontal
    return None, letra_frontal

def leer_historial_ventas():
    with open("ventas.txt", "r") as archivo_ventas:
        for i, linea in enumerate(archivo_ventas):
            datos_venta = linea.strip().split(",")
            if len(datos_venta) < 2:
                continue
            ventas[i] = Venta(datos_venta[0], datos_venta[1])

def leer_revistas():
    with open("revistas.txt", "r") as file:
        for i in range(50):
            linea = file.readline().strip()
            if linea:
                datos = linea.split(",")
                revistas[i] = Revista(datos[0], datos[1])
                letras_frontal, letras_trasera = agregar_letra(revistas[i].letra_frontal, revistas[i].letra_trasera, datos[2])
                revistas[i].letra_frontal = letras_frontal  # Asignar la letra frontal

def main():
    global letras
    letras = None  # Iniciar como None para la lista de letras

    # Insertar letras en la lista enlazada
    for nombre_letra in [
        "Arial12pxRojo", "Times14pxAzul", "Helvetica16pxVerde",
        "Courier10pxNegro", "Verdana18pxNaranja", "ComicSans12pxVioleta",
        "Georgia14pxMarrón", "Arial16pxGris"
    ]:
        letras = insertar_letra_lista(letras, nombre_letra)

    leer_revistas()
    leer_historial_ventas()

    for i in range(50):
        if revistas[i] is None:
            continue
        letra_frontal = revistas[i].letra_frontal
        while letra_frontal is not None:
            letra_actual = letra_frontal.letra  # Aquí se obtiene la letra
            for j in range(50):
                if revistas_contienen_letras[j] is None:
                    continue
                if letras.letra == letra_actual:  # Compara con letra de la lista
                    revistas_contienen_letras[j] = RevistaEncontrada(revistas[i].nombre, revistas[i].pagina, letra_actual)
            letra_frontal, _ = traer_letra(letra_frontal)

    for j in range(50):
        if revistas_contienen_letras[j] is None:
            continue
        for i in range(50):
            if ventas[i] is None:
                continue
            if revistas_contienen_letras[j].nombre == ventas[i].revista_comprada:
                print(f"Posible coincidencia: Cliente {ventas[i].nombre_cliente} compró la revista {ventas[i].revista_comprada}")

if __name__ == "__main__":
    main()
```
