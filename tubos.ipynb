```python
class Nodo:
    def __init__(self, color=None):
        self.color = color
        self.siguiente = None

def agregar_liquido(color, tubo):
    if color is None:
        return tubo
    nuevo_nodo = Nodo(color)
    nuevo_nodo.siguiente = tubo
    return nuevo_nodo

def sacar_liquido(tubo):
    if tubo is not None:
        color_actual = tubo.color
        tubo = tubo.siguiente
        return color_actual, tubo
    return None, tubo

def contar_nodos(tubo):
    contador = 0
    actual = tubo
    while actual is not None:
        contador += 1
        actual = actual.siguiente
    return contador

def leer_tubos():
    tubos = [None] * 10
    try:
        with open("tubos.txt", "r") as file:
            for i in range(10):
                datos = file.readline().strip().split(",")
                for color in datos:
                    tubos[i] = agregar_liquido(color, tubos[i])
    except FileNotFoundError:
        print("El archivo 'tubos.txt' no fue encontrado.")
    imprimir_tubos(tubos)
    return tubos

def imprimir_tubos(tubos):
    for i, tubo in enumerate(tubos):
        actual = tubo
        mensaje = ""
        while actual is not None:
            mensaje += actual.color + ", "
            actual = actual.siguiente
        print(f"Tubo {i+1}: {mensaje.rstrip(', ')}")

def principal():
    tubos = leer_tubos()
    tubos_ordenados = [None] * 10

    for i in range(10):
        while tubos[i] is not None:
            liquido, tubos[i] = sacar_liquido(tubos[i])
            liquido_pasado = False

            for j in range(10):
                if (tubos_ordenados[j] is not None and tubos_ordenados[j].color == liquido 
                        and contar_nodos(tubos_ordenados[j]) < 3):
                    tubos_ordenados[j] = agregar_liquido(liquido, tubos_ordenados[j])
                    liquido_pasado = True
                    break

            if not liquido_pasado:
                for j in range(10):
                    if tubos_ordenados[j] is None:
                        tubos_ordenados[j] = agregar_liquido(liquido, tubos_ordenados[j])
                        break

    imprimir_tubos(tubos_ordenados)

# Ejecutar la funciÃ³n principal
principal()
```
