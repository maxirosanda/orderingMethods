```python
class Letra:
    def __init__(self, letra=None):
        self.letra = letra
        self.siguiente = None

class Revista:
    def __init__(self):
        self.nombre = ""
        self.pagina = ""
        self.letra_frontal = None
        self.letra_trasera = None

class RevistaEncontrada:
    def __init__(self):
        self.nombre = ""
        self.pagina = ""
        self.letra = None

class Venta:
    def __init__(self):
        self.nombre_cliente = ""
        self.revista_comprada = ""

def insertar_letra_lista(lista, letra):
    nuevo_nodo = Letra(letra)
    if lista.letra is None:
        lista.letra = letra
    else:
        nodo_actual = lista
        while nodo_actual.siguiente is not None:
            nodo_actual = nodo_actual.siguiente
        nodo_actual.siguiente = nuevo_nodo

def buscar_lista(lista, posicion):
    nodo_actual = lista
    for i in range(1, posicion + 1):
        if nodo_actual is None:
            return None
        if i == posicion:
            return nodo_actual.letra
        nodo_actual = nodo_actual.siguiente
    return None

def traer_letra(letra_frontal):
    if letra_frontal is not None:
        letra_actual = letra_frontal.letra
        letra_frontal = letra_frontal.siguiente
        return letra_actual
    return None

def agregar_letra(letra_frontal, letra_trasera, letra):
    nueva_letra = Letra(letra)
    if letra_frontal is None:
        letra_frontal = nueva_letra
        letra_trasera = nueva_letra
    else:
        letra_trasera.siguiente = nueva_letra
        letra_trasera = nueva_letra
    return letra_frontal, letra_trasera

def leer_revistas():
    revistas = [Revista() for _ in range(50)]
    with open("revistas.txt", "r") as file:
        for i in range(50):
            linea = file.readline().strip()
            if linea:
                datos = linea.split(",")
                revistas[i].nombre = datos[1]
                revistas[i].pagina = datos[2]
                letra_frontal = letra_trasera = None
                for letra in datos[3:]:
                    letra_frontal, letra_trasera = agregar_letra(letra_frontal, letra_trasera, letra)
                revistas[i].letra_frontal = letra_frontal
    return revistas

def leer_historial_ventas():
    ventas = [Venta() for _ in range(50)]
    with open("ventas.txt", "r") as archivo_ventas:
        for i, linea in enumerate(archivo_ventas):
            datos_venta = linea.strip().split(",")
            ventas[i].nombre_cliente = datos_venta[1]
            ventas[i].revista_comprada = datos_venta[2]
    return ventas

# Inicializar letras y listas
letras = Letra()
for l in ["G", "A", "M", "E", "O", "V", "E", "R"]:
    insertar_letra_lista(letras, l)

revistas = leer_revistas()
revistas_contienen_letras = [RevistaEncontrada() for _ in range(50)]
ventas = leer_historial_ventas()

# Proceso de comparación
for i in range(50):
    while revistas[i].letra_frontal is not None:
        for j in range(1, 9):
            letra = buscar_lista(letras, j)
            if letra == revistas[i].letra_frontal.letra:
                revistas_contienen_letras[j - 1].nombre = revistas[i].nombre
                revistas_contienen_letras[j - 1].pagina = revistas[i].pagina
                revistas_contienen_letras[j - 1].letra = revistas[i].letra_frontal
                break
        revistas[i].letra_frontal = revistas[i].letra_frontal.siguiente

for j in range(8):
    for i in range(50):
        if revistas_contienen_letras[j].nombre == ventas[i].revista_comprada:
            print(f"Posible coincidencia: Cliente {ventas[i].nombre_cliente} compró la revista {ventas[i].revista_comprada}")
```
